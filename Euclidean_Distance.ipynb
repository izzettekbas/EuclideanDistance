```python
#import math
points = [(1,2), (4,6), (6,14), (8,17), (6,10)]
```

```python
def euclideanDistance(point1, point2):
  #distance = math.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)
  distance = ((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)**0.5
  return distance
```

```python
#points_length = len(points)
points_length = 0
for i in points:
  points_length += 1
print(f"Number of points is {points_length}")
```

```python
distances = []
for i in range(points_length):
  for j in range(i+1, points_length):
    distance = euclideanDistance(points[i], points[j])
    distances.append(distance)
    print(f"Distance between {points[i]} and {points[j]} is {distance}")
```

```python
#min_distance = min(distances)
min_distance = distances[0]
for distance in distances:
  if distance < min_distance:
    min_distance = distance
print(f"Minimum distance is {min_distance}")
```
